```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as datetime

df=pd.read_excel(r"C:\Users\91709\OneDrive\Pictures\Documents\Pac_n_mov_customers_data.xlsx")
df
```

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df['Date']=pd.to_datetime(df['Date'],errors='coerce')
df['yearmonth']=df['Date'].dt.to_period('M')
#new dataframe monthly_data
monthly_data=df.groupby('yearmonth').size().reset_index(name='count')

monthly_data['yearmonth']=monthly_data['yearmonth'].dt.to_timestamp()

# Step 8: Plot the result
plt.figure(figsize=(10, 5))
plt.plot(monthly_data['yearmonth'], monthly_data['count'], marker='^', linestyle='-')     
plt.title('Monthly Records Count')
plt.xlabel('Month')
plt.ylabel('Number of Records')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

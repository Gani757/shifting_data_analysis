```python
Q1=df["price"].quantile(0.25)
Q3=df["price"].quantile(0.75)
IQR=Q3-Q1
ll=Q1-(1.5*IQR)
ul=Q3+(1.5*IQR)
oultiers=df[(df["price"]<ll) | (df["price"]>ul)].reset_index(drop=True)
oultiers
```

data cleaning

```python
df.columns
```

```python
df["shifted_from"]=df["shifted_from"].replace({"Drdo":"Drdo Township","Santosh Nagar":"Santhosh Nagar","Saraswathi Colony":"Saraswathi Nagar","Saraswatinagar":"Saraswathi Nagar"})
```

```python
df["shifted_from"]=df["shifted_from"].fillna(df["shifted_from"].mode()[0])
df["shifted_to"]=df["shifted_to"].fillna(df["shifted_to"].mode()[0])
df.isnull().sum()                #treated missing values
```

```python
df.duplicated().sum()
```

```python
#✅ Univariate Analysis (one variable at a time)

sns.countplot(x=df["repeat customer"])
plt.show()
```

```python
sns.violinplot(data=df,x="price",y="repeat customer",hue="repeat customer")
plt.title("Price Distribution by Repeat Customer")
plt.show()
```

observation: Repeat customers tend to spend more and have more diverse pricing patterns.

🟧 Repeat Customers (Yes):
Wider distribution: Prices vary a lot more.

Spread goes from near 0 up to 25,000+.

Likely, repeat customers are getting more premium services, or doing larger/longer shifts.

🔵 One-Time Customers (No):
Tighter spread: Prices mostly between ₹2,000 and ₹10,000.

Distribution is more compact, suggesting smaller or simpler moves.

```python
df["date"]=pd.to_datetime(df["date"],errors="coerce")  #convert date columns to datetime format
df["yearmonth"]=df["date"].dt.to_period("M")
monthly_data=df.groupby("yearmonth")["yearmonth"].agg(count="count").reset_index()

#convert to timestamp for plotting 
monthly_data["yearmonth"]=monthly_data["yearmonth"].dt.to_timestamp()

#sns.relplot(data=monthly_data,x="yearmonth",y="count",kind="line")

# Monthly trend 
plt.plot(monthly_data["yearmonth"],monthly_data["count"])
plt.title("monthly shifting records")
plt.xlabel("month of shifiting")
plt.ylabel("number of shifitings")
plt.xticks(rotation=45)
plt.show()
```
